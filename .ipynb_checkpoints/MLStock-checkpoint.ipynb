```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

from sklearn.linear_model import LinearRegression, ElasticNet, LassoLars
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

df = pd.read_csv('./BTC-USD.csv')
#Convert String to Date
df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d') 
#Set intex in Date
df = df.set_index('Date') 
#Shows Begin History
df.head(10)
```

```python
#Shows End History
df.tail(10)
```

```python
forecast_out = 30
df['Prediction'] = df['Adj Close'].shift(-forecast_out)
x = np.array(df.drop(['Prediction'], 1))
x = x[:-forecast_out]
y = np.array(df['Prediction'])
y = y[:-forecast_out]
#Train the model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
linear_model = LinearRegression()
elasticnet_model = ElasticNet()
lassolars_model = LassoLars()

linear_model.fit(x_train, y_train)
elasticnet_model.fit(x_train, y_train)
lassolars_model.fit(x_train, y_train)
```

```python
linear_model_score = linear_model.score(x_test, y_test)
print('LinearModel score:', linear_model_score)

elasticnet_model_score = elasticnet_model.score(x_test, y_test)
print('ElasticNetModel score:', elasticnet_model_score)

lassolars_model_score = lassolars_model.score(x_test, y_test)
print('LassoLarsModel score:', lassolars_model_score)
```

```python
x_forecast = np.array(df.drop(['Prediction'], 1))[-forecast_out:]
linear_model_forecast_prediction = linear_model.predict(x_forecast)
linear_model_real_prediction = linear_model.predict(np.array(df.drop(['Prediction'], 1)))
elasticnet_model_forecast_prediction = elasticnet_model.predict(x_forecast)
elasticnet_model_real_prediction = elasticnet_model.predict(np.array(df.drop(['Prediction'], 1)))
lassolars_model_forecast_prediction = lassolars_model.predict(x_forecast)
lassolars_model_real_prediction = lassolars_model.predict(np.array(df.drop(['Prediction'], 1)))
```

```python
predicted_dates = []
recent_date = df.index.max()
display_at = 2100
alpha = 0.5

for i in range(forecast_out):
    recent_date += timedelta(days=1)
    predicted_dates.append(recent_date)

#Print Models
plt.plot(df.index[display_at:], linear_model_real_prediction[display_at:], label='Linear Preds', c='green', alpha=alpha)
plt.plot(predicted_dates, linear_model_forecast_prediction, c='green', alpha=alpha)
plt.plot(df.index[display_at:], elasticnet_model_real_prediction[display_at:], label='Elastic Net Preds', c='yellow', alpha=alpha)
plt.plot(predicted_dates, elasticnet_model_forecast_prediction, c='yellow', alpha=alpha)
plt.plot(df.index[display_at:], lassolars_model_real_prediction[display_at:], label='Lasso Lars Preds', c='orange', alpha=alpha)
plt.plot(predicted_dates, lassolars_model_forecast_prediction, c='orange', alpha=alpha)
plt.plot(df.index[display_at:], df['Close'][display_at:], label='Data', c='black', linewidth=3)
plt.legend()
```
